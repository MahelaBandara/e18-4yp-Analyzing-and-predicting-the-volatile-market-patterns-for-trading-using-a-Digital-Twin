```python

# Input
# Input = [[ 1, 2, 3, 4, 5, 6, 7, 8, 9, 0]] -> 1

# Adjusted Input: Expects shape (10,) for an array of 10 integers
tf.random.set_seed(42)
input_layer = tf.keras.layers.Input(shape=(10, 1))
output_layer = tf.keras.layers.Dense(units=10 , activation = 'linear')
sequential_model = tf.keras.Sequential(
    [ 
        tf.keras.layers.Conv1D(filters=256, kernel_size=10, activation='relu'),
        tf.keras.layers.Dense(units=128 , activation = 'relu'),
        tf.keras.layers.Conv1D(filters=64, kernel_size=5, activation='relu', padding='same'),
        tf.keras.layers.PReLU(),
        tf.keras.layers.LSTM(units=64, return_sequences=True),
        tf.keras.layers.Dropout(rate=0.5),
        tf.keras.layers.PReLU(),
        tf.keras.layers.LSTM(units=50),
        tf.keras.layers.Dropout(rate=0.5),
    ]
)

pass_output= sequential_model(input_layer)
output = output_layer(pass_output)

model = tf.keras.models.Model(inputs = input_layer , outputs = output)
model.compile(loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True) , optimizer=tf.keras.optimizers.Adam(1e-9))


#keras tuner daganna
```

```python
model.summary()
```

```python
# Test train - > 60% 30% 10%
from sklearn.model_selection import train_test_split

X = pd.read_csv("previous_values.csv")
Y = pd.read_csv("y_values.csv")

X_train_orig, X_test, Y_train_orig, Y_test = train_test_split(X, Y, test_size=0.1, random_state=42,stratify=Y) 
X_train, X_valid, Y_train, Y_valid = train_test_split(X_train_orig,Y_train_orig, test_size=0.3, random_state=42,stratify=Y_train_orig) 
```

```python
# batch_val = [10, 50, 100, 200, 500, 1000]

# for batch in batch_val:
#     pass
```

```python
history = model.fit(X_train, Y_train, 
          batch_size=1000, 
          verbose=1, 
          epochs=50, 
          validation_data=(X_valid, Y_valid),
          callbacks=[tf.keras.callbacks.EarlyStopping(monitor="val_loss",mode="min" ,patience=3)])
```

```python
model.save('my_model.keras')
```

```python
# loading model
from tensorflow.keras.models import load_model

model_test = load_model('my_model.keras')
```

```python
model_test.summary()
```

```python
model.predict(X_test)
```

```python
values  = tf.nn.softmax(model.predict(X_test))
```

```python
predicted_classes = np.argmax(values , axis = 1)
```

```python
print(predicted_classes)
```

```python
# Assuming your trained model is stored in 'model'
# Assuming the history object is available (captured during training)

# Extract training and validation loss from history
train_loss = history.history['loss']
val_loss = history.history['val_loss']

# Plot the loss curves
plt.plot(train_loss, label='Training Loss')
plt.plot(val_loss, label='Validation Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.show()
```

```python
predictions = predicted_classes

# Extract the last 100 predicted and actual values (assuming enough test data)
last_100_predictions = predictions[-100:]
last_100_actuals = Y_test[-100:]

# Flatten predictions
last_100_predictions_flat = np.ravel(last_100_predictions)

# Create the plot
plt.figure(figsize=(10, 6))  # Adjust figure size as needed
plt.plot(range(len(last_100_actuals)), last_100_actuals, label='Actual Values')
plt.plot(range(len(last_100_predictions_flat)), last_100_predictions_flat, label='Predicted Values')
plt.xlabel('Sample Index (Last 100)')
plt.ylabel('Values')
plt.title('Actual vs. Predicted Values (Last 100)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
from sklearn.metrics import accuracy_score
import numpy as np
# Assuming one-hot encoded labels
accuracy = accuracy_score(Y_test, predicted_classes)
print("Accuracy:", accuracy)

#precision
```
