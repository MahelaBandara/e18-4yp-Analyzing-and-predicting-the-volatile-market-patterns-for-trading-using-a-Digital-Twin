```python
# Convert the 'Date' column to datetime format
data['Date'] = pd.to_datetime(data['Date'])

# Create separate columns for month and date
data['Month'] = data['Date'].dt.month
data['Day'] = data['Date'].dt.day
data['Year'] = data['Date'].dt.year
```

```python
data
```

```python
# Drop rows with empty values
data = data.dropna()
```

```python
data
```

```python
# Define parameters
short_window = 12
long_window = 26
signal_window = 9
rsi_window = 14
cci_window = 20
dx_window = 14
sma_30_window = 30
sma_60_window = 60
bollinger_window = 20
bollinger_std = 2

# Calculate MACD
data['Short_EMA'] = data['BC'].ewm(span=short_window, adjust=False).mean()
data['Long_EMA'] = data['BC'].ewm(span=long_window, adjust=False).mean()
data['MACD'] = data['Short_EMA'] - data['Long_EMA']
data['Signal_Line'] = data['MACD'].ewm(span=signal_window, adjust=False).mean()
data['MACD_Histogram'] = data['MACD'] - data['Signal_Line']

# Calculate Bollinger Bands
data['bollinger_mid'] = data['BC'].rolling(window=bollinger_window).mean()
data['bollinger_std'] = data['BC'].rolling(window=bollinger_window).std()
data['boll_ub'] = data['bollinger_mid'] + (bollinger_std * data['bollinger_std'])
data['boll_lb'] = data['bollinger_mid'] - (bollinger_std * data['bollinger_std'])

# Calculate RSI
delta = data['BC'].diff(1)
gain = delta.where(delta > 0, 0)
loss = -delta.where(delta < 0, 0)
avg_gain = gain.rolling(window=rsi_window, min_periods=1).mean()
avg_loss = loss.rolling(window=rsi_window, min_periods=1).mean()
rs = avg_gain / avg_loss
data['rsi_30'] = 100 - (100 / (1 + rs))

# Calculate CCI (Commodity Channel Index)
typical_price = (data['BC'] + data['BH'] + data['BL']) / 3
sma = typical_price.rolling(window=cci_window).mean()
mean_deviation = (np.abs(typical_price - sma)).rolling(window=cci_window).mean()
data['cci_30'] = (typical_price - sma) / (0.015 * mean_deviation)

# Calculate DX (Directional Movement Index)
delta_x = np.abs(data['BH'] - data['BL'])
delta_y = data['BH'].shift(1) - data['BL']
delta_z = data['BL'] - data['BH'].shift(1)
true_range = pd.concat([delta_x, delta_y, delta_z], axis=1).max(axis=1)
positive_directional_movement = np.where((delta_x > delta_y) & (delta_x > delta_z), delta_x, 0)
negative_directional_movement = np.where((delta_z > delta_x) & (delta_z > delta_y), delta_z, 0)

# Convert arrays to pandas Series
positive_directional_movement = pd.Series(positive_directional_movement, index=data.index)
negative_directional_movement = pd.Series(negative_directional_movement, index=data.index)
true_range = pd.Series(true_range, index=data.index)

adx = (positive_directional_movement.rolling(window=dx_window).mean() / true_range) * 100
di_minus = (negative_directional_movement.rolling(window=dx_window).mean() / true_range) * 100
data['dx_30'] = np.where((adx + di_minus) == 0, 0, np.abs((adx - di_minus) / (adx + di_minus)) * 100)

# Calculate 30-day Simple Moving Average
data['close_30_sma'] = data['BC'].rolling(window=sma_30_window).mean()

# Calculate 60-day Simple Moving Average
data['close_60_sma'] = data['BC'].rolling(window=sma_60_window).mean()

# Drop intermediate columns used for calculations
#data = data.drop(['Short_EMA', 'Long_EMA', 'Signal_Line', 'MACD_Histogram', 'bollinger_mid', 'bollinger_std'], axis=1, errors='ignore')
```

```python
# Assuming your data is in a DataFrame named 'data'
# If 'Date' and 'Time' are already in datetime format, you can skip the conversion
# If not, you may convert 'Date' and 'Time' to datetime using:
# data['Date'] = pd.to_datetime(data['Date'] + ' ' + data['Time'])

# Calculate daily returns
data['Returns'] = data['BC'].pct_change()

# Calculate daily volatility (standard deviation of returns)
data['Volatility'] = data['Returns'].rolling(window=30).std()  # Adjust the window size as needed

# Calculate VIX (Volatility Index)
data['VIX'] = (data['Volatility'] * 100)  # Multiplying by 100 for percentage representation

# Calculate turbulence
data['Turbulence'] = data['Volatility'] * data['Returns']

# Assuming you want daily turbulence, you may want to sum up turbulence over a certain period
data['Turbulence'] = data['Turbulence'].rolling(window=30).sum()  # Adjust the window size as needed

# Drop NaN values resulting from rolling calculations
data = data.dropna()

# Reset index
data = data.reset_index(drop=True)
```
