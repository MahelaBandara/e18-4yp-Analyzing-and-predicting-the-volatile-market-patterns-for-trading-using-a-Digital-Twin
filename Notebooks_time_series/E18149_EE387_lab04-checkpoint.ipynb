```python
import numpy as np
import matplotlib.pyplot as plt
import timeit
```

## TASK 01

```python

# Set the sampling rate and time duration
sampling_rate = 2000  # Hz
duration = 1  # seconds

# Calculate the number of samples
num_samples = int(sampling_rate * duration)

# Set the frequencies, amplitudes, and phases of the sinusoids
frequencies = [2, 5, 9]  # Hz
amplitudes = [4, 3, 0.5]
phases = [0, 0, 0]  # Zero phase for all sinusoids

# Initialize the time and signal arrays
t = np.linspace(0, duration, num_samples)
signal = np.zeros(num_samples)

# Generate the individual sinusoids and add them together
for freq, amp, phase in zip(frequencies, amplitudes, phases):
    sinusoid = amp * np.sin(2 * np.pi * freq * t + phase)
    signal += sinusoid

# Plot the signal against time
plt.figure(figsize=(10, 6))
plt.plot(t, signal)
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.title('Signal')
plt.grid(True)
plt.show()
```

## TASK 07

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.fft import fft, ifft, fftfreq

# Load the data from the CSV file
data = np.loadtxt('wavdata.csv', delimiter=',')

# Calculate the necessary parameters
duration = 10.0
sampling_rate = 1000
num_samples = len(data)
t = np.linspace(0, duration, num_samples)

# Compute the one-sided magnitude spectrum using FFT
fft_spectrum = fft(data)
freqs = fftfreq(num_samples, 1 / sampling_rate)
magnitude_spectrum = np.abs(fft_spectrum) / num_samples

# Apply the band-pass filter
f_min = 320
f_max = 370
indices = np.logical_or(freqs < f_min, freqs > f_max)
fft_spectrum_filtered = fft_spectrum.copy()
fft_spectrum_filtered[indices] = 0

# Compute the inverse FFT to obtain the filtered signal
filtered_signal = ifft(fft_spectrum_filtered).real

# Plot the filtered signal in the time domain
plt.figure(figsize = (12, 6))
plt.plot(t, filtered_signal)
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.grid(True)
plt.show()
```

```python
from scipy.fft import fft, fftfreq

# Load the data from the CSV file
data = np.loadtxt('wavdata.csv', delimiter=',')

# Calculate the necessary parameters
duration = 10.0  
sampling_rate = 1000
num_samples = len(data)
t = np.linspace(0, duration, num_samples)

# Compute the one-sided magnitude spectrum of the original signal using FFT
fft_spectrum_original = fft(data)
freqs = fftfreq(num_samples, 1 / sampling_rate)
magnitude_spectrum_original = np.abs(fft_spectrum_original) / num_samples

# Apply the band-pass filter to the FFT spectrum
freq_min = 320
freq_max = 370
indices = np.logical_or(freqs < freq_min, freqs > freq_max)
fft_spectrum_filtered = fft_spectrum_original.copy()
fft_spectrum_filtered[indices] = 0

# Compute the one-sided magnitude spectrum of the filtered signal using the modified FFT spectrum
magnitude_spectrum_filtered = np.abs(fft_spectrum_filtered) / num_samples

# Plot the Graphs 
plt.figure(figsize=(15,13))

# original signal
plt.subplot(2,1,1)
plt.plot(freqs[:num_samples // 2], magnitude_spectrum_original[:num_samples // 2])
plt.xlabel('Frequency (Hz)')
plt.ylabel('Magnitude Spectrum')
plt.title('Original Signal')
plt.grid(True)

# filtered signal
plt.subplot(2,1,2)
plt.plot(freqs[:num_samples // 2], magnitude_spectrum_filtered[:num_samples // 2])
plt.xlabel('Frequency (Hz)')
plt.ylabel('Magnitude Spectrum')
plt.title('Filtered Signal')
plt.grid(True)

plt.show()
```
